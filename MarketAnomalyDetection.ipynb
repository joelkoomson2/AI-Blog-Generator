```python
# Import IsolationForest
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report
import numpy as np

# Prepare features (X)
X = df[['VIX']]

# Create and train Isolation Forest Model
iso_forest = IsolationForest(
    n_estimators=1000,  # Number of trees
    contamination="auto", # Expected proportion of outliers (10%)
    random_state=42,
    warm_start=True,
    n_jobs=-1 # Use all CPU cores
)

# Fit and predict
# predict() returns 1 for inliers, -1 for outliers
predictions = iso_forest.fit_predict(X)

# Convert predictions to binary format (0 for normal, 1 for anomaly)
# This makes it consistent with your original labels
y_pred = np.where(predictions == -1, 1, 0)

# Print classification report
print("Classification Report:")
print(classification_report(y, y_pred))

# Calculate and print anomaly scores
anomaly_scores = iso_forest.score_samples(X)
print("\nAnomaly Score Statistics:")
print(f"Mean score: {np.mean(anomaly_scores):.4f}")
print(f"Min score: {np.min(anomaly_scores):.4f}")
print(f"Max score: {np.max(anomaly_scores):.4f}")
```

```python
import pandas as pd

def create_financial_time_split(
    df: pd.DataFrame,
    train_years: int = 4,
    validation_weeks: int = 52,
    test_weeks: int = 52
):
    """
    Specialized split for financial time series.

    Args:
        df : pd.DataFrame
            DataFrame with weekly data. Must have a 'Data' column.
        train_years : int, default 4
            Number of years for training data.
        validation_weeks : int, default 52
            Number of weeks for validation data.
        test_weeks : int, default 52
            Number of weeks for test data.

    Returns:
        Tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]
            train_data, validation_data, test_data
    """
    # Convert 'Data' column to datetime
    df['Data'] = pd.to_datetime(df['Data'])
    df = df.sort_values('Data').reset_index(drop=True)

    # Calculate required sizes
    train_size = train_years * 52  # 52 weeks per year
    total_required_size = train_size + validation_weeks + test_weeks

    if len(df) < total_required_size:
        raise ValueError(
            f"Not enough data: required {total_required_size} rows, "
            f"but got {len(df)}"
        )

    # Create splits
    train_end = train_size
    val_end = train_end + validation_weeks

    train_data = df.iloc[:train_end].copy()
    validation_data = df.iloc[train_end:val_end].copy()
    test_data = df.iloc[val_end:val_end + test_weeks].copy()

    # Verify splits
    print(f"Training period:   {train_data['Data'].min().date()} to {train_data['Data'].max().date()}")
    print(f"Validation period: {validation_data['Data'].min().date()} to {validation_data['Data'].max().date()}")
    print(f"Testing period:    {test_data['Data'].min().date()} to {test_data['Data'].max().date()}")

    return train_data, validation_data, test_data


if __name__ == "__main__":
    # Load your uploaded CSV
    df = pd.read_csv('FinancialMarketData.csv')

    # Run the split
    train_data, val_data, test_data = create_financial_time_split(df)

    # Optionally save the splits
    train_data.to_csv('train_data.csv', index=False)
    val_data.to_csv('validation_data.csv', index=False)
    test_data.to_csv('test_data.csv', index=False)

    print("Splits saved: train_data.csv, validation_data.csv, test_data.csv")
```
